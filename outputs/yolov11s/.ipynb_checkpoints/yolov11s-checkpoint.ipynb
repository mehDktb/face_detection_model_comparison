```python
import cv2
import torch
import time
from ultralytics import YOLO
device ="cpu"
model = YOLO("yolov11s-face.pt").to(device) 


video_path = '/home/mehdi_ktb/Documents/Face_Detection/emkan_sanji/face_detection/output.mp4'  
cap = cv2.VideoCapture(video_path)

frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
out = cv2.VideoWriter('output_video.mp4', fourcc, fps, (frame_width, frame_height))


prev_time = 0
frame_count = 0  
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

  
    if frame_count % 1 != 0:
        frame_count += 1
        continue

    curr_time = time.time()
    fps_text = f'FPS: {1 / (curr_time - prev_time):.2f}'
    prev_time = curr_time

    results = model.predict(frame, conf=0.40, verbose=False, device=device)

    for result in results[0].boxes:  
        x1, y1, x2, y2 = result.xyxy[0]  
        conf = result.conf[0] 
        cls = result.cls[0]  


        if conf >= 0.60:
            x1, y1, x2, y2 = map(int, [x1, y1, x2, y2])  # Convert to integer
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, f'{conf:.2f}', (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.putText(frame, fps_text, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

    out.write(frame)

    cv2.imshow('Face Detection', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

    frame_count += 1

cap.release()
out.release()
cv2.destroyAllWindows()
```
